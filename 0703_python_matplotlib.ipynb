Matplotlib의 pyplot 모듈을 plt라는 이름으로 불러오고, 이제 plt는 MATLAB 스타일의 인터페이스를 제공하며, 그래프를 그리는 함수들이 모여있다.

```python
import matplotlib.pyplot as plt
import numpy as np

#np.array([])로 1차원 배열들을 만들고
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

#plt.plot() 함수로 x,y축에 그려낸다.
plt.plot(xpoints, ypoints)
plt.show()
```

plt.plot(x,y, 'o') 세번째 인자를 주면 양 끝지점에 점만 찍힌다

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

fmt 문법, '마커:점선:색상' 순으로 옵션을 줄수있다.
- '-' solid<br>
':'	dotted<br>
'--'	dashed<br>
'-.'	dashdot<br>
선 제거 None<br>
- 'r'	Red
'g'	Green
'b'	Blue
'c'	Cyan
'm'	Magenta
'y'	Yellow
'k'	Black
'w'	White

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

ms = 마커사이즈, mec = 마커엣지컬러, mfc = 마커페이스컬러 순으로 수정가능.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

마커색상은 헥사코드나, 색상이름 문자열도 가능.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
plt.show()
```

위에서 적었던 선 스타일은 ls = 'dashed' 등으로 줄수있고, c = '색상'으로 선 색상도 바꿀수 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted', color = 'b')
plt.show()
```

선 두께는 linewidth = lw 로 수정가능.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

plt.plot()함수를 각각 써주어 한 화면에 나타냈다.<br>
plt.xlabel("") 라벨과 plt.title("") 제목도 붙혀준다.<br>
fontdict = {'글씨체':'', '색상':'', '사이즈':''}로 제목과 라벨 폰트를 수정할수있다.<br>
plt.grid() 그리드는 격자.

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(y1)
plt.plot(y2)

plt.grid(axis='y')

plt.show()
```

x1,x2, y1,y2 를 모두 따로 정의했고, plt.plot(x축, y축) 순서만 맞추면 된다<br>
plt.grid(color, ls, lw) 격자도 색상을 바꿀수있다.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```

title의 위치를 location = 'left' 등으로 옮길수있다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'left')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

plt.subplot(행, 열, n번째plot) 으로 몇개의 plot창을 띄울지 결정가능<br>
- plt.title("")은 항상 plt.plot()이후마다!
- plt.suptitle("")은 항상 subplot이 다 끝나고 plt.show()직전에 슈퍼타이틀로 붙힌다!

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

plt.subplot(행, 열, n번째plot) 똑같이 사용했지만, plot창이 2행으로 보인다.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```
